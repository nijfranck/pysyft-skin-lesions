```python
# Import libraries
import os, os.path
from glob import glob
from PIL import Image
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
```

```python
directory = '../skin-lesions/images/'

image_path = glob(os.path.join(directory, '*.jpg'))
imageid = {os.path.splitext(os.path.basename(x))[0]: x for x in image_path}
print(len(imageid))
```

```python
lesion_type = {
    'MEL': 'melanoma',
    'NV': 'melanocytic nevi',
    'BCC': 'basal cell carcinoma',
    'AK': 'Actinic keratoses',
    'BKL': 'benign keratosis',
    'DF': 'dermatofibroima',
    'VASC': 'vascular lesions',
    'SCC': 'Squamous cell carcinom',
    'UNK': 'Keratosis pilaris'
}
```

```python
data = pd.read_csv('../skin-lesions/table.csv')
data.head()
```

```python
lesions_data = data.loc[data['image'].isin(imageid)]
lesions_data.head()
```

```python
lesions_data['path'] = lesions_data['image'].map(imageid.get)
```
